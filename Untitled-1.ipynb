```python
# PHẦN 2: phân tích Cohort
# tạo 
new_store_df2 = store_df[['Customer ID','Order Date', 'Revenue' ]].copy()
new_store_df2['Order Date'] = pd.to_datetime (store_df['Order Date'])   
new_store_df2 = new_store_df2[new_store_df2['Order Date'].dt.year == 2018].reset_index()

# tạo min df
df_min_day= new_store_df2.groupby(['Customer ID'])['Order Date'].min().reset_index()
df_min_day.rename(columns={'Order Date':'min_order_month'}, inplace= True)
df_min_day
```

```python

cohort_day= pd.merge(new_store_df2,df_min_day, on=['Customer ID'])

cohort_day['cohort_index'] = cohort_day['Order Date'].dt.month - cohort_day['min_order_month'].dt.month +1
cohort_day['Order Date'] = cohort_day['Order Date'] + pd.offsets.MonthBegin(-1)
cohort_day['min_order_month'] = cohort_day['min_order_month'] + pd.offsets.MonthBegin(-1)
cohort_day
```

```python


cohort_data_final = cohort_day.groupby(['min_order_month', 'Order Date', 'cohort_index'])['Customer ID'].count().reset_index()

cohort_data_final2 = cohort_day.groupby(['min_order_month', 'Order Date', 'cohort_index'])['Revenue'].sum().reset_index()
cohort_pivot = cohort_data_final.pivot_table(index='min_order_month', columns='cohort_index', values='Customer ID')
cohort_pivot2 = cohort_data_final2.pivot_table(index='min_order_month', columns='cohort_index', values='Revenue')
```
